# import essential librarys

```python
from pyomo.environ import *
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import random
from colorama import Fore, Back, Style
import logging
logging.getLogger('pyomo.core').setLevel(logging.FATAL)
```

# Case 1: base model 

```python
# Create a model
model=AbstractModel(name="Economic Dipatch")
# Parameters
model.Demand=Param(mutable=True,initialize=100)
model.Costwind=Param(mutable=True,initialize=20)
model.Costsolar=Param(mutable=True,initialize=25)
model.Coststorage=Param(mutable=True,initialize=20)
# Decision Variables
model.Pwind=Var(bounds=(0,40),within=NonNegativeReals,initialize=0)
model.Psolar=Var(bounds=(0,50),within=NonNegativeReals,initialize=0)
model.Pstorage=Var(bounds=(0,10),within=NonNegativeReals,initialize=0)
# Constraints
model.energy_balance=Constraint(expr=model.Pwind+model.Psolar+model.Pstorage==model.Demand)
# Objective function
model.Obj=Objective(expr=model.Costwind*model.Pwind+model.Costsolar*model.Psolar
                    +model.Coststorage*model.Pstorage,sense=minimize)
```

```python
# Specifing Solver
opt=SolverFactory('glpk')
instance=model.create_instance()

#solve & update the instance
results=opt.solve(instance)
print(results)
```

```python
print("Psolar=",value(instance.Psolar))
print("Pstorage=",value(instance.Pstorage))
print("Pwind=",value(instance.Pwind))
print("OF=", value(instance.Obj))
```

# Case 2: Thermal Units & Carbon emissions minimization problem  
In this case, we add 2 thermal generating units with different generation costs  to the problem to make it more comprehensive. In addition, we increase the capacity of renewable sources and the amount of network Demand. By considering thermal units, we can further investigate the impact of Carbon emissions in Economic Dispatch problem.

```python
# reading data from .csv file
df=pd.read_csv("..//Data//Genetion Data.csv")
df.head(6)
```

### Scenario  1: simple ED with the objective of minimizing the Total cost, considering Termal Units.

```python
# Create a model
model=AbstractModel()
model.i=RangeSet(len(df))
# Parameters
model.Demand=Param(mutable=True,default=300,within=NonNegativeReals)

# Decision Variables
def P_bound (model,i):
    return (df.loc[i-1,"Pmin"],df.loc[i-1,"Pmax"])
model.P=Var(model.i,bounds=P_bound,within=NonNegativeReals,initialize=0)

# Constraints
def energy_balance(model):
    return sum(model.P[i] for i in model.i)==model.Demand
model.energy_balance=Constraint(rule= energy_balance)

# Objective function
def obj(model):
    return sum(df.loc[i-1,"a"]*model.P[i]**2+df.loc[i-1,"b"]*model.P[i]+df.loc[i-1,"c"] for i in model.i)

model.Obj=Objective(rule=obj,sense=minimize)
```

```python
# Specifing Solver
opt=SolverFactory('ipopt')
instance=model.create_instance()
instance.Demand=400
#solve & update the instance
results=opt.solve(instance)
print(results)
```

```python
for ii in instance.i:
    print(df.loc[ii-1,"Name"]+"\t=", value(instance.P[ii]),"KW")
print("="*50)
print("Total Generation=\t",round(value(instance.energy_balance),5),"KW","\nTotal Demand=\t",value(instance.Demand),"KW")   
print("OF=", value(instance.Obj))
```

### Scenario  2: ED with the objective of minimizing the Total cost & Co2 Emission, considering Termal Units.

```python
## the model is the same as before, we just need to define an objective function for Co2 Emission

# Create a model
model=AbstractModel()
model.i=RangeSet(len(df))
# Parameters
model.Demand=Param(mutable=True,default=350,within=NonNegativeReals)
# monetizing Parameter for carbon emissions.
model.mu=Param(initialize=1,within=Reals,mutable=True)

# Decision Variables
def P_bound (model,i):
    return (df.loc[i-1,"Pmin"],df.loc[i-1,"Pmax"])
model.P=Var(model.i,bounds=P_bound,within=NonNegativeReals,initialize=0)

# Constraints
def energy_balance(model):
    return sum(model.P[i] for i in model.i)==model.Demand
model.energy_balance=Constraint(rule= energy_balance)
# Objective function
def Cost_obj(model):
    return sum(df.loc[i-1,"a"]*model.P[i]**2+df.loc[i-1,"b"]*model.P[i]+df.loc[i-1,"c"] for i in model.i)

def Emission_obj(model):
    return sum(df.iloc[i-1,4]*model.P[i] for i in model.i)


def Total_obj(model):
    return model.mu*Emission_obj(model)+Cost_obj(model)

model.ObjTotal=Objective(rule=Total_obj,sense=minimize)
```

```python
# Specifing Solver
opt=SolverFactory('ipopt')
instance=model.create_instance()
instance.mu=0.1
instance.Demand=400
#solve & update the instance
results=opt.solve(instance)
# print(results)
```

```python
for ii in instance.i:
    print(df.loc[ii-1,"Name"]+"\t=", value(instance.P[ii]),"KW")
print("="*50)
print("Total Generation=\t",round(value(instance.energy_balance),5),"KW","\nTotal Demand=\t",value(instance.Demand),"KW")   
print("OF=", value(instance.ObjTotal))
```

# Case 3: ED problem considering Demand/Wind/Solar variability
In this case, we assume that Demand/Wind/Solar change over 24h horizon with a known Profile. Like Case 2 our objective is to minimize the Total generation cost & Co2 emission. Ramp up/down constraints are also considered for Termal units. 

```python
# reading data from .csv files

Profile_df=pd.read_csv("..//Data//Daily_Profile.csv")
Profile_df.head(6)

for i in range(0,len(Profile_df.columns)):
    plt.figure(figsize=(8,8))
    Profile_df.iloc[:,i].plot(kind='bar',label=Profile_df.columns[i])
    plt.xlabel('Time (h)', fontweight ='bold', fontsize = 10) 
    plt.ylabel('Variability (%)', fontweight ='bold', fontsize = 10) 
    plt.legend()
```

```python
## the model is the same as before, we just need to consider Time variation and define Ramp up/down Constraints
# Create a model
model=AbstractModel()
model.i=RangeSet(len(df))
model.t=RangeSet(len(Profile_df))
# Parameters
model.Demand=Param(mutable=True,default=350,within=NonNegativeReals)
# monetizing Parameter for carbon emissions.
model.mu=Param(initialize=0,within=Reals,mutable=True)

# Decision Variables
def P_bound (model,i,t):
    if(df.loc[i-1,"Name"]=="Wind"):
        return (df.loc[i-1,"Pmin"],df.loc[i-1,"Pmax"]*Profile_df.loc[t-1,"Wind"])
    elif(df.loc[i-1,"Name"]=="Solar"):
        return (df.loc[i-1,"Pmin"],df.loc[i-1,"Pmax"]*Profile_df.loc[t-1,"Solar"])
    else: 
        return (df.loc[i-1,"Pmin"],df.loc[i-1,"Pmax"])
model.P=Var(model.i,model.t,bounds=P_bound,within=NonNegativeReals,initialize=0)

# Constraints
def energy_balance(model,t):
    return sum(model.P[i,t] for i in model.i)==model.Demand*Profile_df.loc[t-1,"Demand"]
model.energy_balance=Constraint(model.t,rule= energy_balance)

def Ramp_up(model,i,t):
    if(df.loc[i-1,"Name"]=="G1" or df.loc[i-1,"Name"]=="G2"):
        if(t<=23):
            return model.P[i,t+1]-model.P[i,t]<=0.2*df.loc[i-1,"Pmax"]
        else:
            return Constraint.Skip
    else:
        return Constraint.Skip
model.Ramp_up=Constraint(model.i,model.t,rule= Ramp_up)

def Ramp_down(model,i,t):
    if(df.loc[i-1,"Name"]=="G1" or df.loc[i-1,"Name"]=="G2"):
        if(t<=23):
            return model.P[i,t]-model.P[i,t+1]<=0.20*df.loc[i-1,"Pmax"]
        else:
            return Constraint.Skip
    else:
        return Constraint.Skip
model.Ramp_down=Constraint(model.i,model.t,rule= Ramp_down)

# Objective function
def Cost_obj(model):
    return sum(df.loc[i-1,"a"]*model.P[i,t]**2+df.loc[i-1,"b"]*model.P[i,t]+df.loc[i-1,"c"] for i in model.i for t in model.t)

def Emission_obj(model):
    return sum(df.iloc[i-1,4]*model.P[i,t] for i in model.i for t in model.t)


def Total_obj(model):
    return model.mu*Emission_obj(model)+Cost_obj(model)

model.ObjTotal=Objective(rule=Total_obj,sense=minimize)
```

```python
# Specifing Solver
opt=SolverFactory('ipopt')
instance=model.create_instance()
instance.mu=0.1
instance.Demand=330
#solve & update the instance
results=opt.solve(instance)
# print(results)
results.write()
```

```python
Data_hour={}
for t in instance.t:
    print( "Hour",t,"="*50)
    data=[]
    for ii in instance.i:
        data.append(round(value(instance.P[ii,t]),2))
        print(df.loc[ii-1,"Name"]+"\t=", value(instance.P[ii,t]),"KW")
    print("Total Generation=\t",round(value(instance.energy_balance[t]),5),"KW","\nTotal Demand=\t",
      value(instance.Demand)*Profile_df.loc[t-1,"Demand"],"KW")
    data.append(round(value(instance.energy_balance[t]),2))
    data.append(round(value(instance.Demand)*Profile_df.loc[t-1,"Demand"],2))
    Data_hour[t]=data
    if (round(value(instance.energy_balance[t]))!=round(value(instance.Demand)*Profile_df.loc[t-1,"Demand"])):
        print(Back.RED + 'infeasible')
        print(Style.RESET_ALL)

print("="*25,"end","="*25)
   
print("OF=", value(instance.ObjTotal))
```

```python
Curtailment_dict={}
for ii in (1,2):
    Curtailment_dict[df.loc[ii-1,"Name"]]=[round(df.loc[ii-1,"Pmax"]*Profile_df.loc[t-1,df.loc[ii-1,"Name"]]
                                           -value(instance.P[ii,t],2)) for t in instance.t]

for key in Curtailment_dict.keys():
    print(key+"_Curtailment",":",Curtailment_dict[key],'\n')
```

```python
## save result 
df_data =pd.DataFrame.from_dict(Data_hour,orient='index',columns=['Wind', 'Solar', 'Storage', 'G1','G2','Total_Generation', 'Total_Demand'])
df_data.to_excel("Case3_result.xlsx")
df_data.head()
```

# Case 4: ED problem considering storage degradation, efficiency loss
In this case, we assume that a storage with maximum capacity of 150 Kwh, discharge/charge rate of 50KWh and 0.95 efficiency is available. by using a storage over 24h horizon we expect to reduce the renewable unit's curtailment. Like previous case our objective is to minimize the Total generation cost & Co2 emission.  

```python
# reading data from .csv file
df_2=pd.read_csv("..//Data//Genetion Data_noESS.csv")
df_2.head(6)
```

```python
## the model is the same as before, we just need to consider Storage output variation over time and define Storage Constraints
# Create a model
model=AbstractModel()
model.i=RangeSet(len(df_2))
model.t=RangeSet(len(Profile_df))
## Parameters
model.Demand=Param(mutable=True,default=350,within=NonNegativeReals)
# monetizing Parameter for carbon emissions.
model.mu=Param(initialize=0,within=Reals,mutable=True)

# Storage efficiency factor
model.gamma=Param(initialize=0.95,mutable=True)

# Decision Variables
def P_bound (model,i,t):
    if(df_2.loc[i-1,"Name"]=="Wind"):
        return (df_2.loc[i-1,"Pmin"],df_2.loc[i-1,"Pmax"]*Profile_df.loc[t-1,"Wind"])
    elif(df_2.loc[i-1,"Name"]=="Solar"):
        return (df_2.loc[i-1,"Pmin"],df_2.loc[i-1,"Pmax"]*Profile_df.loc[t-1,"Solar"])
    else: 
        return (df_2.loc[i-1,"Pmin"],df_2.loc[i-1,"Pmax"])
model.P=Var(model.i,model.t,bounds=P_bound,within=NonNegativeReals,initialize=0)

# Storage Dcharge/Charge & SOC Variable
model.SOC=Var(model.t,bounds=(10,150),within=Reals,initialize=0)
model.Pch=Var(model.t,bounds=(0,50),within=Reals,initialize=0)
model.Pdch=Var(model.t,bounds=(0,50),within=Reals,initialize=0)


# Constraints
def energy_balance(model,t):
    return sum(model.P[i,t] for i in model.i)+model.Pdch[t]-model.Pch[t]==model.Demand*Profile_df.loc[t-1,"Demand"]
model.energy_balance=Constraint(model.t,rule= energy_balance)

def Ramp_up(model,i,t):
    if(df_2.loc[i-1,"Name"]=="G1" or df_2.loc[i-1,"Name"]=="G2"):
        if(t<=23):
            return model.P[i,t+1]-model.P[i,t]<=0.2*df_2.loc[i-1,"Pmax"]
        else:
            return Constraint.Skip
    else:
        return Constraint.Skip
model.Ramp_up=Constraint(model.i,model.t,rule= Ramp_up)

def Ramp_down(model,i,t):
    if(df_2.loc[i-1,"Name"]=="G1" or df_2.loc[i-1,"Name"]=="G2"):
        if(t<=23):
            return model.P[i,t]-model.P[i,t+1]<=0.20*df_2.loc[i-1,"Pmax"]
        else:
            return Constraint.Skip
    else:
        return Constraint.Skip
model.Ramp_down=Constraint(model.i,model.t,rule= Ramp_down)

# storage Constraints
def rule_SOC(model,t):
        if(t==1):
            return model.SOC[t]== 10+model.gamma*model.Pch[t]-model.Pdch[t]/model.gamma
        else:
            return model.SOC[t]== model.SOC[t-1]+model.gamma*model.Pch[t]-model.Pdch[t]/model.gamma
            
    
model.Csoc=Constraint(model.t,rule= rule_SOC)


# Objective function
def Cost_obj(model):
    return 20*sum(model.Pdch[t]-model.Pch[t] for t in model.t)+sum(df_2.loc[i-1,"a"]*model.P[i,t]**2+df_2.loc[i-1,"b"]*model.P[i,t]+df_2.loc[i-1,"c"] for i in model.i for t in model.t)

def Emission_obj(model):
    return sum(df_2.iloc[i-1,4]*model.P[i,t] for i in model.i for t in model.t)


def Total_obj(model):
    return model.mu*Emission_obj(model)+Cost_obj(model)

model.ObjTotal=Objective(rule=Total_obj,sense=minimize)
```

```python
# Specifing Solver
opt=SolverFactory('ipopt')
instance=model.create_instance()
instance.mu=0.1
instance.Demand=305
#solve & update the instance
results=opt.solve(instance)
# print(results)
results.write()
```

```python
P_ESS=[value(instance.Pdch[t]-instance.Pch[t]) for t in instance.t]
Data_hour={}
for t in instance.t:
    print( "Hour",t,"="*50)
    data=[]
    for ii in instance.i:
        data.append(round(value(instance.P[ii,t]),2))
        print(df_2.loc[ii-1,"Name"]+"\t=", value(instance.P[ii,t]),"KW")
    print("Total Generation=\t",round(value(instance.energy_balance[t]),5),"KW","\nTotal Demand=\t",
      value(instance.Demand)*Profile_df.loc[t-1,"Demand"],"KW")
    data.append(round(value(instance.energy_balance[t]),2))
    data.append(round(value(instance.Demand)*Profile_df.loc[t-1,"Demand"],2))
    Data_hour[t]=data
    if (round(value(instance.energy_balance[t]))!=round(value(instance.Demand)*Profile_df.loc[t-1,"Demand"])):
        print(Back.RED + 'infeasible')
        print(Style.RESET_ALL)

print("="*25,"end","="*25)
   
print("OF=", value(instance.ObjTotal))
```

# Case 5: ED problem considering network and power flow constraints
In this case, we consider a three bus network, and assume All the network Demand is located in Bus-2. Also, Solar and ESS is located in Bus-2, Generator-1 is located in Bus-1, Generator-2 and Wind is located in Bus-3. The power flow equation is based on DC approximation to avoid further complexity. Like previous case our objective is to minimize the Total generation cost & Co2 emission.  


```python
# reading data from .csv file
Profile_df=pd.read_csv("..//Data//Daily_Profile.csv")
df_2=pd.read_csv("..//Data//Genetion Data_noESS.csv")
Branch_df=pd.read_excel("..//Data//branch.xlsx")

Branch_df.head()
```

```python
model=AbstractModel()
model.bus=RangeSet(len(Branch_df))
model.node=RangeSet(len(Branch_df))
model.t=RangeSet(len(Profile_df))
model.gen=RangeSet(len(df_2))

## Parameters
model.Demand=Param(mutable=True,default=350,within=NonNegativeReals)
model.Sbase=Param(mutable=True,initialize=100)
slackbus=3

# monetizing Parameter for carbon emissions.
model.mu=Param(initialize=0,within=Reals,mutable=True)

# Storage efficiency factor
model.gamma=Param(initialize=0.95,mutable=True)

def P_bound (model,i,t):
    if(df_2.loc[i-1,"Name"]=="Wind"):
        return (df_2.loc[i-1,"Pmin"]/model.Sbase,df_2.loc[i-1,"Pmax"]*Profile_df.loc[t-1,"Wind"]/model.Sbase)
    elif(df_2.loc[i-1,"Name"]=="Solar"):
        return (df_2.loc[i-1,"Pmin"]/model.Sbase,df_2.loc[i-1,"Pmax"]*Profile_df.loc[t-1,"Solar"]/model.Sbase)
    else: 
        return (df_2.loc[i-1,"Pmin"]/model.Sbase,df_2.loc[i-1,"Pmax"]/model.Sbase)    
    
model.P=Var(model.gen,model.t,bounds=P_bound,within=NonNegativeReals,initialize=0)

def Pij_bound (model,bus,node,t):
    if ((Branch_df['From'] == bus) & (Branch_df['To'] == node)).any():
        return (-Branch_df.loc[bus-1,"Limit"]/model.Sbase,Branch_df.loc[bus-1,"Limit"]/model.Sbase)
    elif ((Branch_df['From'] == node) & (Branch_df['To'] == bus)).any():
        return (-Branch_df.loc[bus-1,"Limit"]/model.Sbase,Branch_df.loc[bus-1,"Limit"]/model.Sbase)
    else :
        return (0,0)
    
model.Pij=Var(model.bus,model.node,model.t,bounds=Pij_bound,within=Reals,initialize=0)
model.delta=Var(model.bus,model.t,bounds=[-np.pi,np.pi],within=Reals,initialize=0)

# Storage Dcharge/Charge & SOC Variable
model.SOC=Var(model.t,bounds=(0.1,1.5),within=Reals,initialize=0)
model.Pch=Var(model.t,bounds=(0,0.5),within=Reals,initialize=0)
model.Pdch=Var(model.t,bounds=(0,0.5),within=Reals,initialize=0)

# Constraints
def Ramp_up(model,i,t):
    if(df_2.loc[i-1,"Name"]=="G1" or df_2.loc[i-1,"Name"]=="G2"):
        if(t<=23):
            return model.P[i,t+1]-model.P[i,t]<=0.2*df_2.loc[i-1,"Pmax"]/model.Sbase
        else:
            return Constraint.Skip
    else:
        return Constraint.Skip
model.Ramp_up=Constraint(model.gen,model.t,rule= Ramp_up)

def Ramp_down(model,i,t):
    if(df_2.loc[i-1,"Name"]=="G1" or df_2.loc[i-1,"Name"]=="G2"):
        if(t<=23):
            return model.P[i,t]-model.P[i,t+1]<=0.20*df_2.loc[i-1,"Pmax"]/model.Sbase
        else:
            return Constraint.Skip
    else:
        return Constraint.Skip
model.Ramp_down=Constraint(model.gen,model.t,rule= Ramp_down)

def Line_flow(model,bus,node,t):
    if ((Branch_df['From'] == bus) & (Branch_df['To'] == node)).any():
        X_value=Branch_df.loc[(Branch_df['From'] == bus) & (Branch_df['To'] == node),'X'].values[0]
        return model.Pij[bus,node,t]== 1/X_value*(model.delta[bus,t]-model.delta[node,t])
    elif ((Branch_df['From'] == node) & (Branch_df['To'] == bus)).any():
        X_value=Branch_df.loc[(Branch_df['From'] == node) & (Branch_df['To'] == bus),'X'].values[0]
        return model.Pij[bus,node,t]== 1/X_value*(model.delta[bus,t]-model.delta[node,t])
    else:
        return Constraint.Skip
model.flow=Constraint(model.bus,model.node,model.t,rule= Line_flow)

def energybalance(model,bus,t):
    if bus==1:
        return model.P[3,t]==model.Pij[1,2,t]+model.Pij[1,3,t]
    elif bus==2:
        return model.P[2,t]+model.Pdch[t]-model.Pch[t]==model.Pij[2,1,t]+model.Pij[2,3,t]+model.Demand*Profile_df.loc[t-1,"Demand"]/model.Sbase
    else:
        return model.P[4,t]+model.P[1,t]==model.Pij[3,1,t]+model.Pij[3,2,t]
model.energybalance=Constraint(model.bus,model.t,rule= energybalance)

# storage Constraints
def rule_SOC(model,t):
        if(t==1):
            return model.SOC[t]== 0.1+model.gamma*model.Pch[t]-model.Pdch[t]/model.gamma
        else:
            return model.SOC[t]== model.SOC[t-1]+model.gamma*model.Pch[t]-model.Pdch[t]/model.gamma   
model.Csoc=Constraint(model.t,rule= rule_SOC)

# Objective function
def Cost_obj(model):
    return 20*sum((model.Pdch[t]-model.Pch[t])*model.Sbase for t in model.t)+sum(df_2.loc[i-1,"a"]*model.P[i,t]*model.Sbase**2+df_2.loc[i-1,"b"]*model.P[i,t]*model.Sbase+df_2.loc[i-1,"c"] for i in model.gen for t in model.t)

def Emission_obj(model):
    return sum(df_2.iloc[i-1,4]*model.P[i,t]*model.Sbase for i in model.gen for t in model.t)


def Total_obj(model):
    return model.mu*Emission_obj(model)+Cost_obj(model)

model.ObjTotal=Objective(rule=Total_obj,sense=minimize)
```

```python
# Specifing Solver
opt=SolverFactory('ipopt')
instance=model.create_instance()
instance.mu=0.1
instance.Demand=305

for t in instance.t:
    instance.delta[slackbus,t].fix(0)
    
#solve & update the instance
results=opt.solve(instance)
# print(results)
results.write()
```
